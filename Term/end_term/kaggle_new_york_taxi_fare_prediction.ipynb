## Import the package

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from zipfile import ZipFile
```

## Import the data sets.

```python
data_path = os.getcwd()
train_data_path = data_path + '/data/taxi_fare_train.csv'
test_data_path = data_path + '/data/taxi_fare_test.csv'

df_train = pd.read_csv(train_data_path, nrows=500000)
df_test = pd.read_csv(test_data_path)
```

```python
df_train.info()
```

```python
df_train.head()
```

```python
train_Y = np.array(df_train['fare_amount'])
ids = df_test['key']
```

```python
df_train = df_train.drop(['key', 'fare_amount'] , axis=1)
df_test = df_test.drop(['key'] , axis=1)
df = pd.concat([df_train,df_test])
df.head()
df['distance_2D'] = ((df['dropoff_longitude'] - df['pickup_longitude'])**2 + (df['dropoff_latitude'] - df['pickup_latitude'])**2)**0.5
df = df.drop(['pickup_longitude', 'pickup_latitude', 'dropoff_longitude', 'dropoff_latitude', 'pickup_datetime'], axis=1)
df.head()
```

```python
df['passenger_count'] = df['passenger_count'].fillna(-1)
df['distance_2D'] = df['distance_2D'].fillna(-1)
```

## Prediction

#### With the Gradient Boosting Machine Model

```python
from sklearn import datasets, metrics
from sklearn.ensemble import GradientBoostingRegressor
```

```python
train_num = train_Y.shape[0]
train_X = df[:train_num]
test_X = df[train_num:]

reg = GradientBoostingRegressor()
reg.fit(train_X, train_Y)
pred = reg.predict(test_X)
```

## Output the Results

```python
data_path = os.getcwd()
data_path += "/data/my_result/"
pred = np.array(pred)
sub = pd.DataFrame({'key': ids, 'fare_amount': pred})
sub.to_csv(data_path + 'taxi_1111_finalexam.csv', index=False) 
```
